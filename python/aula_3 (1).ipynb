```python
import math

y_sin = []

x = list ([x/100 for x in range(0,2000)])

for value in x:
  y_sin.append(math.sin(value))
```

Visualização

```python
import matplotlib.pyplot as plt

linel, = plt.plot(x, y_sin)

plt.margins(0.20)

plt.xlabel('X', color = 'g')
plt.ylabel('Y', color = 'g')

plt.title('seno')
linel.set_color('g')

plt.show()
```

```python
y_sin = []
y_cos = []
y_tan = []

x = list ([x/100 for x in range(0,2000)])

for value in x:
  y_sin.append(math.sin(value))
  y_cos.append(math.cos(value))
  y_tan.append(math.tan(value))
```

```python
line1, = plt.plot(x, y_sin)
line2, = plt.plot(x, y_cos)

plt.margins(0.20)

plt.xlabel('X', color = 'g')
plt.ylabel('Y', color = 'g')

plt.title('seno')
linel.set_color('g')

plt.legend([line1, line2], ['seno','cosseno'])

plt.show()
```

```python
line1, = plt.plot(x, y_sin)
line2, = plt.plot(x, y_cos)
line3, = plt.plot(x, y_tan)

plt.margins(0.20)

plt.xlabel('X', color = 'g')
plt.ylabel('Y', color = 'g')

plt.title('seno')
linel.set_color('g')

plt.legend([line1, line2, line3], ['seno','cosseno', 'tan'])

plt.show()
```

```python
fig = plt.figure(figsize=(13,15))
ax = fig.add_subplot(111, projection = '3d')

line1, = ax.plot(x, w, y_sin)
line2, = ax.plot(x, w, y_cos)
line3, = ax.plot(x, w, y_tan)

ax.set_xlabel('X', c = 'r')
ax.set_ylabel('W', c = 'r')
ax.set_zlabel('Y', c = 'r')

plt.title("teste 3d")
plt.legend([line1,line2,line3], ['seno', 'cosseno','tangente'])
```

```python
y_sin = []
y_cos = []
y_tan = []

x = list ([x/100 for x in range(0,2000)])

for value in x:
  y_sin.append(math.sin(value))
  y_cos.append(math.cos(value))
  y_tan.append(math.tan(value) * 0.005)
```

```python
df = pd.concat([
    pd.DataFrame({'X':x, 'Y':y_sin, 'function':['seno']*len(y_sin)}),
    pd.DataFrame({'X':x, 'Y':y_cos, 'function':['cosseno']*len(y_cos)}),
    pd.DataFrame({'X':x, 'Y':y_tan, 'function':['tangente']*len(y_tan)}),
])
```

```python
fig = px.line(df, x='X', y='Y', color = 'function')
fig.show()
```
